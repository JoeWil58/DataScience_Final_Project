This notebook represents our main analysis and computation while performing clustering on historic baseball data. We attempted to cluster teams by performance metrics into canonical eras that professional baseball is often split into.  
We implemented a K-means algorithm from scratch by hand. We also created a Gaussian Mixture Model to show how soft clustering may perform differently in this context.

```python
from pybaseball.lahman import *
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

#### K-Means by Hand

```python
Teams = teams()

#Runs per game
Teams['R_per_G'] = Teams['R'] / Teams['G']

#Homeruns per game
Teams['HR_per_G'] = Teams['HR'] / Teams['G']

#Hits per game
Teams['H_per_G'] = Teams['H'] / Teams['G']

#Pitcher strikeouts per game
Teams['SO_p_per_G'] = Teams['SOA'] / Teams['G']

#Batter strikeouts per game
Teams['SO_b_per_G'] = Teams['SO'] / Teams['G']

#Proportion of games that were complete games
Teams['CG_prop'] = Teams['CG'] / Teams['G']

# Clean Team dataframe based on these columns
Teams = Teams.dropna(subset=['R_per_G','HR_per_G', 'H_per_G', 'SO_p_per_G', 'SO_b_per_G', 'CG_prop'])

# Label observations based on true era
Teams['Era'] = np.array(['blue']*Teams.shape[0])
Teams.loc[Teams['yearID'] <= 1900,'Era'] = 0
Teams.loc[(1901 <= Teams['yearID']) & (Teams['yearID'] <= 1919),'Era'] = 1
Teams.loc[(1920 <= Teams['yearID']) & (Teams['yearID'] <= 1947),'Era'] = 2
Teams.loc[(1948 <= Teams['yearID']) & (Teams['yearID'] <= 1960),'Era'] = 3
Teams.loc[(1961 <= Teams['yearID']) & (Teams['yearID'] <= 1977),'Era'] = 4
Teams.loc[(1978 <= Teams['yearID']) & (Teams['yearID'] <= 1995),'Era'] = 5
Teams.loc[1996 <= Teams['yearID'],'Era'] = 6

# Create dataframe which is subset of full Team dataframe
Notable_features = Teams[['R_per_G','HR_per_G', 'H_per_G', 'SO_p_per_G', 'SO_b_per_G', 'CG_prop']]
Notable_features = Notable_features.dropna().copy()

Teams
```

```python
def dist(x1, x2):
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
# Set number of clusters
K = 7

Notable_features_list = np.array(Notable_features)

# Randomly select K centroids for initialization
centroids_idx = np.random.choice(list(range(len(Notable_features_list))),K)
centroids = [Notable_features_list[i] for i in centroids_idx]

# Find number of observations and initialize distance list and clusters array
shape = len(Notable_features_list)
dists = [0]*K

# Will hold the classifications 
clusters = np.array([-1]*shape)

# Loop through observations and assign them to initiali clusters
for i in range(shape):
    dists = [dist(Notable_features_list[i],cent) for cent in centroids]
    assignment = dists.index(min(dists))
    clusters[i] = assignment

# Create dictionary of integer to color mapping for coloring eras
colors = {0: 'red', 1: 'orange', 2: 'grey', 3: 'green', 4: 'blue', 5:'purple', 6:'pink'}
cluster_colors = [colors[i] for i in clusters]
    
# Loop through data 100 times and adjust clusters each time
for temp in range(100):
    
    # Loop through data and calculate distances to assign clusters
    for i in range(shape):
        
        # Calculate distance from data point to each cluster
        dists = [dist(Notable_features_list[i],cent) for cent in centroids]
        
        # Select minimum distance of these and append the index of that to the cluster list
        assignment = dists.index(min(dists))
        
        # Assign this cluster to the node in the assignment list
        clusters[i] = assignment
        
    # Loop through centroids and adjust centroids based on mean of points in cluster
    for j in range(K):
        
        centroids[j]=[np.sum(Notable_features_list[clusters==j,0])/len(Notable_features_list[clusters==j,0]),
                      np.sum(Notable_features_list[clusters==j,1])/len(Notable_features_list[clusters==j,1]),
                      np.sum(Notable_features_list[clusters==j,2])/len(Notable_features_list[clusters==j,2]),
                      np.sum(Notable_features_list[clusters==j,3])/len(Notable_features_list[clusters==j,3]),
                      np.sum(Notable_features_list[clusters==j,4])/len(Notable_features_list[clusters==j,4]),
                      np.sum(Notable_features_list[clusters==j,5])/len(Notable_features_list[clusters==j,5])]
```

#### K-Means with SciKit-Learn

```python
#import the relevant data
#this dataset will include some NaN values, so it's 'dirty'
Teams_dirty = teams()

#Showcase the structure of the dataframe
Teams_dirty
```

```python
#All of the available features out of the box.
#Most of these will not be relevant to us.
Teams.columns
```

```python
#clean the dataset
#Only need to drop rows where there are NaN's
#in the columns that we care about
Teams = Teams_dirty.dropna(subset=['G', 'R', 'HR', 'H', 'SOA', 'SO', 'CG', 'BB', '2B', '3B']).copy()
```

```python
#We can start by calculating our Lahman features for the 'Teams' dataset  

#Runs per game
Teams['R_per_G'] = Teams['R'] / Teams['G']

#Homeruns per game
Teams['HR_per_G'] = Teams['HR'] / Teams['G']

#Hits per game
Teams['H_per_G'] = Teams['H'] / Teams['G']

#Batter strikeouts per game
Teams['SO_b_per_G'] = Teams['SO'] / Teams['G']

#Proportion of games that were complete games
Teams['CG_prop'] = Teams['CG'] / Teams['G']

#Batter walks per game
Teams['BB_per_G'] = Teams['BB'] / Teams['G']

#Singles (used to calculate slugging)
Teams['1B'] = Teams['H'] - (Teams['2B'] + Teams['3B'] + Teams['HR'])

#Team slugging percentage
#SLG keeps track of a team's propensity towards extra base hits
#by dividing the total number of bases a team gets by the number
#of at-bats they take
Teams['SLG'] = (Teams['1B'] + (2*Teams['2B']) + (3*Teams['3B']) + (4*Teams['HR'])) / Teams['AB']
```

```python
#Create column containing true era information

#The year boundaries for all eras
conditions = [
    (Teams['yearID'] < 1900),
    (Teams['yearID'] >= 1900) & (Teams['yearID'] < 1919),
    (Teams['yearID'] >= 1919) & (Teams['yearID'] < 1947),
    (Teams['yearID'] >= 1947) & (Teams['yearID'] < 1960),
    (Teams['yearID'] >= 1960) & (Teams['yearID'] < 1977),
    (Teams['yearID'] >= 1977) & (Teams['yearID'] < 1995),
    (Teams['yearID'] >= 1995)
]

#The corresponding era choices
choices = ['19th Century', 'Dead-Ball', 'Live-Ball', 'Integration', 'Expansion', 'Free Agency', 'Long-Ball']
Teams['time_era'] = np.select(conditions, choices, default='?')
```

```python
#Check to see if our new columns are behaving as expected
Teams
```

```python
#Create an array of features from our newly created metrics
feat_array = Teams.values[:,[48, 49, 50, 51, 52, 53, 55]]
```

```python
#Import sklearn's KMeans to check against our
#hand-implemented algorithm
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7, random_state=42).fit(feat_array)
```

```python
#This plots the cluster assigned to a team by its corresponding year
years = Teams['yearID']
labels = kmeans.labels_
Teams['cluster_label'] = labels
fig,ax = plt.subplots(figsize=(25,10))
ax.scatter(years, labels, alpha=0.3)
```

As we can see, each cluster has a density of points located somehwere along the year line. If this clustering were perfect, we would see a tight group of points for each cluster centered around different years corresponding to the true eras. Obviously, this result is quite far from that ideal.

```python
#The code below creates lists of data for each cluster
#This data is used to create box and whisker plots for each cluster

c_0 = list(Teams['yearID'].loc[Teams['cluster_label'] == 0])
c_1 = list(Teams['yearID'].loc[Teams['cluster_label'] == 1])
c_2 = list(Teams['yearID'].loc[Teams['cluster_label'] == 2])
c_3 = list(Teams['yearID'].loc[Teams['cluster_label'] == 3])
c_4 = list(Teams['yearID'].loc[Teams['cluster_label'] == 4])
c_5 = list(Teams['yearID'].loc[Teams['cluster_label'] == 5])
c_6 = list(Teams['yearID'].loc[Teams['cluster_label'] == 6])
#cluster_box = [c_0, c_1, c_2, c_3, c_4, c_5, c_6]
#cluster_box = [c_4, c_3, c_2, c_0, c_5, c_6, c_1]
cluster_box = [c_1, c_6, c_5, c_0, c_2, c_3, c_4]

fig,ax=plt.subplots(figsize=(25,10))
ax.boxplot(cluster_box, vert=0)

#plot vertical era boundaries
ax.axvline(1900, linestyle='--', color='gray', alpha=0.5)
ax.axvline(1920, linestyle='--', color='gray', alpha=0.5)
ax.axvline(1947, linestyle='--', color='gray', alpha=0.5)
ax.axvline(1960, linestyle='--', color='gray', alpha=0.5)
ax.axvline(1977, linestyle='--', color='gray', alpha=0.5)
ax.axvline(1995, linestyle='--', color='gray', alpha=0.5)

#pinpoint a few interesting points
ax.scatter([1884,1884,1884], [1,2,3], color='red', s=300)
ax.text(1877, 7.55, '19th Century', fontsize=13)
ax.text(1907, 7.55, 'Dead-Ball', fontsize=13)
ax.text(1931, 7.55, 'Live-Ball', fontsize=13)
ax.text(1949.5, 7.55, 'Integration', fontsize=13)
ax.text(1964.5, 7.55, 'Expansion', fontsize=13)
ax.text(1981, 7.55, 'Free Agency', fontsize=13)
ax.text(2007, 7.55, 'Long-Ball', fontsize=13)
ax.set_xlabel('Year', fontsize=17)
ax.set_ylabel('Cluster', fontsize=17)
ax.set_title('Boxplots by Cluster', y=1.07, fontsize=20)
print()
```

This plot, which will be referenced in our presentation, shows how the clusters are composed. As we can see, there is significant overlap between clusters in terms of the years in which they span. However, it is promising that they do all have distinct means. The top cluster is the only one that has a very tight spread. The three red points represent interesting points since they are 19 century teams that get clustered with teams from the 70s through present day. These teams, as is discovered below, are mostly clustered down here because they had very high batter strikeout rates for their time.

```python
#Find the two teams that were clustered with modern day teams but are major outliers
Teams.loc[(Teams['yearID'] == 1884) & (Teams['cluster_label'] == 1)]
```

```python
#This code calculates how many standard deviations away from the 
#average 19th century team these two teams are in terms of
#strikeouts by batters per game

mean_so_19 = np.mean(Teams['SO_b_per_G'].loc[Teams['time_era'] == '19th Century'])
std_so_19 = np.std(Teams['SO_b_per_G'].loc[Teams['time_era'] == '19th Century'])

(Teams.loc[(Teams['yearID'] == 1884) & (Teams['cluster_label'] == 1)]['SO_b_per_G'] - mean_so_19) / std_so_19
```

As can be seen above, these two teams have a strikeout per game value (for batters) that is over 3 standard deviations away from the average 19th century value.

```python
#Find index of 19th century teams to use as x axis
idx_19 = list(Teams.loc[Teams['time_era'] == '19th Century'].index)

#Use strikeouts per game as y axis
so_19 = Teams.loc[Teams['time_era'] == '19th Century']['SO_b_per_G']

fig,ax = plt.subplots(figsize=(10,8))
ax.scatter(idx_19, so_19)
ax.scatter([130, 154], [7.220183, 7.235294], color='red', label='Points over 3 std dev from mean')

#plot mean line
ax.axhline(np.mean(so_19), color='coral', label='Mean')

#plot 2 std deviations line above and below mean
ax.axhline(np.mean(so_19) + (2*np.std(so_19)), linestyle='--', color='coral')
ax.axhline(np.mean(so_19) - (2*np.std(so_19)), linestyle='--', color='coral')

ax.text(300, 5.5, '2 Std Deviations')
ax.text(300, 0.37, '2 Std Deviations')
ax.set_xlabel('19 Century Team Index')
ax.set_ylabel('Strikeouts (Batters) Per Game')
ax.set_title('Strikeouts Per Game for 19th Century Teams')
ax.legend()
```

This plot simply highlights the 2 interesting points found above. It also shows that there are many points that are more than 2 standard deviations away from the mean for strikeouts per game for 19th century teams. The two red points just happen to be particularly egregious.

#### Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
```

```python
Teams = teams()

#Runs per game
Teams['R_per_G'] = Teams['R'] / Teams['G']

#Homeruns per game
Teams['HR_per_G'] = Teams['HR'] / Teams['G']

#Hits per game
Teams['H_per_G'] = Teams['H'] / Teams['G']

#Pitcher strikeouts per game
Teams['SO_p_per_G'] = Teams['SOA'] / Teams['G']

#Batter strikeouts per game
Teams['SO_b_per_G'] = Teams['SO'] / Teams['G']

#Proportion of games that were complete games
Teams['CG_prop'] = Teams['CG'] / Teams['G']

# Clean Team dataframe based on these columns
Teams = Teams.dropna(subset=['R_per_G','HR_per_G', 'H_per_G', 'SO_p_per_G', 'SO_b_per_G', 'CG_prop'])

# Label observations based on true era
Teams['Era'] = np.array(['blue']*Teams.shape[0])
Teams.loc[Teams['yearID'] <= 1900,'Era'] = 0
Teams.loc[(1901 <= Teams['yearID']) & (Teams['yearID'] <= 1919),'Era'] = 1
Teams.loc[(1920 <= Teams['yearID']) & (Teams['yearID'] <= 1947),'Era'] = 2
Teams.loc[(1948 <= Teams['yearID']) & (Teams['yearID'] <= 1960),'Era'] = 3
Teams.loc[(1961 <= Teams['yearID']) & (Teams['yearID'] <= 1977),'Era'] = 4
Teams.loc[(1978 <= Teams['yearID']) & (Teams['yearID'] <= 1995),'Era'] = 5
Teams.loc[1996 <= Teams['yearID'],'Era'] = 6

# Create dataframe which is subset of full Team dataframe
Notable_features = Teams[['R_per_G','HR_per_G', 'H_per_G', 'SO_p_per_G', 'SO_b_per_G', 'CG_prop']]
Notable_features = Notable_features.dropna().copy()
```

```python
# Create GMM Model
GMM = GaussianMixture(n_components=7, covariance_type='full')

# Fit model on our full dataset
GMM.fit(Notable_features)

# Predict on full dataset to find cluster
cluster_assignments = GMM.predict(Notable_features)

# Plot predictions
fig, ax = plt.subplots(figsize=(20,10))
colors = {0: 'red', 1: 'orange', 2: 'grey', 3: 'green', 4: 'blue', 5:'purple', 6:'pink'}
cluster_colors = [colors[i] for i in cluster_assignments]
colors_to_map = ['red', 'orange', 'grey', 'green', 'blue', 'purple', 'pink']
ax.scatter(Teams['R_per_G'], Teams['SO_p_per_G'], c=cluster_colors, cmap=matplotlib.colors.ListedColormap(colors_to_map), alpha=0.8)
ax.set_ylabel('Strikeouts Pitcher per Game')
ax.set_xlabel('Runs Per Game')
ax.set_title('Strikeouts Pitcher per Game vs Runs per Game (Predicted)')
```
